```python
import pandas as pd
import numpy as np
from tabulate import tabulate

# 🔹 Mostrar título con estilo
print("\n📊 **Análisis Bivariado: DEFINITIVA vs SEXO** 📊\n")

# 🔹 Agrupar por SEXO y calcular estadísticas básicas
df2 = df.groupby("SEXO")["DEFINITIVA"].agg(["mean", "std", "var", "count"]).reset_index()

# 🔹 Calcular el coeficiente de variación (CV) en porcentaje
df2["Coeficiente de Variación (%)"] = round((df2["std"] / df2["mean"]) * 100, 1)

# 🔹 Calcular el error estándar de la media (std / sqrt(n))
df2["Error Estándar"] = round(df2["std"] / np.sqrt(df2["count"]), 3)

# 🔹 Calcular el Intervalo de Confianza del 95% para la media
df2["IC (95%)"] = df2.apply(
    lambda row: (round(row["mean"] - 1.96 * row["Error Estándar"], 2),
                 round(row["mean"] + 1.96 * row["Error Estándar"], 2)), axis=1)

# 🔹 Renombrar columnas para mayor claridad
df2.columns = ["SEXO", "Media", "Desviación Estándar", "Varianza", "Cantidad",
               "Coef. Variación (%)", "Error Estándar", "IC (95%)"]

# 🔹 Mostrar tabla en formato bonito
print(tabulate(df2, headers="keys", tablefmt="fancy_grid", showindex=False))
```

#**Análisis Bivariado: DEFINITIVA vs PROGRAMA**

```python
import pandas as pd
import numpy as np
from scipy import stats
from tabulate import tabulate

# 🔹 Mostrar título con estilo
print("\n📊 **Análisis Bivariado: DEFINITIVA vs PROGRAMA** 📊\n")

# 🔹 Agrupar por PROGRAMA y calcular estadísticas básicas
df2 = df.groupby("PROGRAMA")["DEFINITIVA"].agg(["mean", "std", "var", "count"]).reset_index()

# 🔹 Calcular el coeficiente de variación (CV) en porcentaje
df2["Coeficiente de Variación (%)"] = round((df2["std"] / df2["mean"]) * 100, 1)

# 🔹 Calcular el error estándar de la media (std / sqrt(n))
df2["Error Estándar"] = round(df2["std"] / np.sqrt(df2["count"]), 3)

# 🔹 Valor crítico t (para cada programa según su tamaño de muestra)
df2["Valor Crítico t"] = df2["count"].apply(lambda n: round(stats.t.ppf(0.975, df=n-1), 3))

# 🔹 Intervalo de Confianza del 95% con distribución T
df2["IC_T (95%)"] = df2.apply(
    lambda row: (round(row["mean"] - row["Valor Crítico t"] * row["Error Estándar"], 2),
                 round(row["mean"] + row["Valor Crítico t"] * row["Error Estándar"], 2)), axis=1)

# 🔹 Intervalo de Confianza del 95% con distribución Z (Aproximación Normal)
Z_value = 1.96  # Para 95%
df2["IC_Z (95%)"] = df2.apply(
    lambda row: (round(row["mean"] - Z_value * row["Error Estándar"], 2),
                 round(row["mean"] + Z_value * row["Error Estándar"], 2)), axis=1)

# 🔹 Renombrar columnas para mayor claridad
df2.columns = ["PROGRAMA", "Media", "Desviación Estándar", "Varianza", "Cantidad",
               "Coef. Variación (%)", "Error Estándar", "Valor Crítico t", "IC_T (95%)", "IC_Z (95%)"]

# 🔹 Mostrar tabla en formato bonito
print(tabulate(df2, headers="keys", tablefmt="fancy_grid", showindex=False))

```
