```python
# Importar librer√≠as
import pandas as pd

# Lista de edades
edades = [
    19,18,18,21,18,18,18,17,18,22,18,18,18,18,18,22,21,19,18,19,18,19,18,17,
    19,20,19,18,17,18,20,19,20,22,18,19,19,18,19,20,19,18,17,19,19,19,19,18,
    17,18,19,20,23,18,18,20,18,18,18,18,18,19,17,17,20,19,18,19,19,21,19,21,
    17,19,18,18,17,17,18,18,17,18,19,17,18,18,18,18,20,18,18,18,17,19,18,17,
    18,18,18,18,18,18,18,18,19,18,19,18,18,17,18,18,18,18,18,18,18,18,24,18,
    19,19,18,17,17,17,20,17,18,17,18,18,19,18,18,18
]

# Crear DataFrame
df = pd.DataFrame(edades, columns=["Edad"])

# Calcular medidas de tendencia central
media = df["Edad"].mean()
mediana = df["Edad"].median()
moda = df["Edad"].mode()[0]

# Mostrar resultados
print("1. Medidas de tendencia central para datos individuales:")
print(f"Media: {media:.2f}")
print(f"Mediana: {mediana}")
print(f"Moda: {moda}")
```

# **ùüê.  ùêáùêöùê•ùê•ùêû ùê¶ùêûùêùùê¢ùêùùêöùê¨ ùêùùêû ùê©ùê®ùê¨ùê¢ùêúùê¢√≥ùêß ùê©ùêöùê´ùêö ùêùùêöùê≠ùê®ùê¨ ùê¢ùêßùêùùê¢ùêØùê¢ùêùùêÆùêöùê•ùêûùê¨: ùêêùüè, ùêêùüë, ùêÉùüë, ùêÉùüñ ùêû ùê¢ùêßùê≠ùêûùê´ùê©ùê´ùêûùê≠ùêû ùê•ùê®ùê¨ ùêØùêöùê•ùê®ùê´ùêûùê¨ ùêûùêßùêúùê®ùêßùê≠ùê´ùêöùêùùê®ùê¨**

```python
# Calcular medidas de posici√≥n
q1 = df["Edad"].quantile(0.25)  # Primer cuartil
q3 = df["Edad"].quantile(0.75)  # Tercer cuartil
d3 = df["Edad"].quantile(0.3)   # Tercer decil
d8 = df["Edad"].quantile(0.8)   # Octavo decil

# Imprimir resultados e interpretaci√≥n
print("2. Medidas de posici√≥n para datos individuales:\n")

print(f"Q1 (Primer Cuartil)  = {q1}")
print(f"Q3 (Tercer Cuartil)  = {q3}")
print(f"D3 (Tercer Decil)    = {d3}")
print(f"D8 (Octavo Decil)    = {d8}")
```

# **ùüë. ùêÇùêöùê•ùêúùêÆùê•ùêû ùêûùê• ùêÉùê¢ùêöùê†ùê´ùêöùê¶ùêö ùêùùêû ùêÇùêöùê£ùêö ùêùùêû ùêúùêöùê£ùêö ùê≤ ùêõùê¢ùê†ùê®ùê≠ùêûùê¨ ùêû ùê¢ùêßùê≠ùêûùê´ùê©ùê´ùêûùê≠ùêû ùê•ùê®ùê¨ ùêØùêöùê•ùê®ùê´ùêûùê¨ ùêöùê≠√≠ùê©ùê¢ùêúùê®ùê¨ ùê¨ùê¢ ùê•ùê®ùê¨ ùê°ùêöùê≤**

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


# Calcular Q1, Q3 e IQR
q1 = np.percentile(edades, 25)
q3 = np.percentile(edades, 75)
iqr = q3 - q1

# L√≠mites para detectar at√≠picos
limite_inferior = q1 - 1.5 * iqr
limite_superior = q3 + 1.5 * iqr

# Detectar valores at√≠picos
valores_atipicos = [x for x in edades if x < limite_inferior or x > limite_superior]

# Mostrar el diagrama de caja
plt.figure(figsize=(10, 2))
sns.boxplot(x=edades, color='lightblue')
plt.title(" Diagrama de Caja y Bigotes - Edades / Sofia Jaramillo")
plt.xlabel("Edad")
plt.grid(axis='x')
plt.show()

# Imprimir resultados
print(" Valores at√≠picos detectados:")
if valores_atipicos:
    print(sorted(set(valores_atipicos)))
else:
    print("No se detectaron valores at√≠picos.")
```

**Los valores 21, 22, 23 y 24 a√±os se encuentran fuera del rango normal de edades del grupo.**

Estos datos se consideran at√≠picos porque superan el l√≠mite superior del rango intercuart√≠licO, lo que indica que est√°n significativamente por encima de la mayor√≠a de las edades (que se concentran entre los 17 y 20 a√±os).

# **ùüí. ùêÇùê®ùêßùê¨ùê≠ùê´ùêÆùê≤ùêö ùê•ùêö ùêìùêöùêõùê•ùêö ùêöùê†ùê´ùêÆùê©ùêöùêùùêö ùêùùêû ùêüùê´ùêûùêúùêÆùêûùêßùêúùê¢ùêöùê¨ ùê≠ùêöùê• ùêúùê®ùê¶ùê® ùê¨ùêû ùê¢ùêßùêùùê¢ùêúùêö ùêûùêß ùêûùê• ùêïùêàùêÉùêÑùêé ùüè**

```python
import pandas as pd
import numpy as np
import math

# Usamos los valores de edad ya proporcionados anteriormente
df = pd.DataFrame(edades, columns=["Edad"])

# Paso 1: N√∫mero de clases (Regla de Sturges)
n = len(df)
k = round(1 + 3.322 * math.log10(n))

# Paso 2: Rango y amplitud
rango = df["Edad"].max() - df["Edad"].min()
amplitud = math.ceil(rango / k)

# Paso 3: Crear intervalos
limite_inferior = df["Edad"].min()
bins = [limite_inferior + i * amplitud for i in range(k + 1)]

# Paso 4: Cortar datos en intervalos
intervalos = pd.cut(df["Edad"], bins=bins, right=False)

# Paso 5: Calcular frecuencias
tabla = intervalos.value_counts().sort_index().reset_index()
tabla.columns = ["Intervalo", "Frecuencia"]

# Frecuencia acumulada
tabla["Frec. Acumulada"] = tabla["Frecuencia"].cumsum()

# Frecuencia relativa
tabla["Frec. Relativa"] = (tabla["Frecuencia"] / n).round(2)

# Frecuencia relativa acumulada
tabla["Frec. Rel. Acum."] = tabla["Frec. Relativa"].cumsum().round(2)

# Mostrar tabla
print("Tabla de Frecuencias Agrupadas:")
print(tabla.to_string(index=False))
```

# **ùüì. ùêÇùê®ùêßùê¨ùê≠ùê´ùêÆùê≤ùêö ùêûùê• ùêáùê¢ùê¨ùê≠ùê®ùê†ùê´ùêöùê¶ùêö ùêùùêû ùêüùê´ùêûùêúùêÆùêûùêßùêúùê¢ùêöùê¨ ùêöùêõùê¨ùê®ùê•ùêÆùê≠ùêöùê¨ ùê≤ ùêùùê¢ùêõùêÆùê£ùêû ùêûùê• ùê©ùê®ùê•√≠ùê†ùê®ùêßùê®**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

# Usamos los datos de edades proporcionados anteriormente
df = pd.DataFrame(edades, columns=["Edad"])

# N√∫mero de clases (Sturges)
n = len(df)
k = round(1 + 3.322 * math.log10(n))

# Rango y amplitud
rango = df["Edad"].max() - df["Edad"].min()
amplitud = math.ceil(rango / k)

# Bins
limite_inferior = df["Edad"].min()
bins = [limite_inferior + i * amplitud for i in range(k + 1)]

# Intervalos y tabla
frecuencias = pd.cut(df["Edad"], bins=bins, right=False)
tabla = frecuencias.value_counts().sort_index().reset_index()
tabla.columns = ["Intervalo", "Frecuencia"]
tabla["Marca de clase"] = [(interval.left + interval.right) / 2 for interval in tabla["Intervalo"]]

# Gr√°fico
plt.figure(figsize=(10, 6))

# Histograma: color lavanda
plt.bar(tabla["Marca de clase"], tabla["Frecuencia"],
        width=amplitud, align='center', alpha=0.8, edgecolor='black',
        color='#cdb4db', label="Histograma")

# Pol√≠gono: azul oscuro con puntos azul cielo
plt.plot(tabla["Marca de clase"], tabla["Frecuencia"],
         marker='o', color='#3a0ca3', linewidth=2, label="Pol√≠gono")
plt.scatter(tabla["Marca de clase"], tabla["Frecuencia"], color='#4cc9f0', zorder=5)

# Est√©tica del gr√°fico
plt.title("Histograma y Pol√≠gono de Frecuencias - Sofia Jaramillo", fontsize=14)
plt.xlabel("Edad", fontsize=12)
plt.ylabel("Frecuencia", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.4)
plt.legend()
plt.tight_layout()
plt.show()
```

# **ùüî. ùêÇùê®ùêßùê¨ùê≠ùê´ùêÆùê≤ùêö ùêûùê• ùêáùê¢ùê¨ùê≠ùê®ùê†ùê´ùêöùê¶ùêö ùêùùêû ùêüùê´ùêûùêúùêÆùêûùêßùêúùê¢ùêöùê¨ ùê´ùêûùê•ùêöùê≠ùê¢ùêØùêöùê¨ ùê≤ ùê©ùê®ùê•√≠ùê†ùê®ùêßùê®**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

# Usamos los datos de edades ya proporcionados
df = pd.DataFrame(edades, columns=["Edad"])

# N√∫mero de clases (Sturges)
n = len(df)
k = round(1 + 3.322 * math.log10(n))

# Rango y amplitud
rango = df["Edad"].max() - df["Edad"].min()
amplitud = math.ceil(rango / k)

# Bins
limite_inferior = df["Edad"].min()
bins = [limite_inferior + i * amplitud for i in range(k + 1)]

# Intervalos
frecuencias = pd.cut(df["Edad"], bins=bins, right=False)
tabla = frecuencias.value_counts().sort_index().reset_index()
tabla.columns = ["Intervalo", "Frecuencia"]

# Marca de clase y frecuencia relativa
tabla["Marca de clase"] = [(interval.left + interval.right) / 2 for interval in tabla["Intervalo"]]
tabla["Frecuencia Relativa"] = tabla["Frecuencia"] / n

# Gr√°fico
plt.figure(figsize=(10, 6))

# Histograma: verde salvia suave
plt.bar(tabla["Marca de clase"], tabla["Frecuencia Relativa"],
        width=amplitud, align='center', alpha=0.95,
        edgecolor='#6B4226', color='#B5CDA3', label="Histograma Relativo")

# Pol√≠gono: rosa viejo con puntos pastel
plt.plot(tabla["Marca de clase"], tabla["Frecuencia Relativa"],
         marker='o', color='#A26769', linewidth=2.5, label="Pol√≠gono Relativo")
plt.scatter(tabla["Marca de clase"], tabla["Frecuencia Relativa"], color='#E4BAD4', s=80, zorder=5)

# Est√©tica
plt.title("Histograma y Pol√≠gono de Frecuencias Relativas", fontsize=15, color="#4B3F3F")
plt.xlabel("Edad", fontsize=12)
plt.ylabel("Frecuencia Relativa", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

# **ùüï. ùêÇùê®ùêßùê¨ùê≠ùê´ùêÆùê≤ùêö ùêûùê• ùêáùê¢ùê¨ùê≠ùê®ùê†ùê´ùêöùê¶ùêö ùêùùêû ùêüùê´ùêûùêúùêÆùêûùêßùêúùê¢ùêöùê¨ ùêöùêúùêÆùê¶ùêÆùê•ùêöùêùùêöùê¨ ùê≤ ùê©ùê®ùê•√≠ùê†ùê®ùêßùê®**

# **ùüñ. ùêåùêûùêùùê¢ùêùùêöùê¨ ùêùùêû ùê≠ùêûùêßùêùùêûùêßùêúùê¢ùêö ùêúùêûùêßùê≠ùê´ùêöùê• ùêöùê†ùê´ùêÆùê©ùêöùêùùêöùê¨: ùêåùêûùêùùê¢ùêö, ùêåùêûùêùùê¢ùêöùêßùêö ùê≤ ùêåùê®ùêùùêö**

```python
import pandas as pd
import numpy as np
import math

# Usamos la lista de edades proporcionada anteriormente
df = pd.DataFrame(edades, columns=["Edad"])

# Paso 1: Calcular n√∫mero de clases (Sturges)
n = len(df)
k = round(1 + 3.322 * math.log10(n))

# Paso 2: Calcular el rango y la amplitud
rango = df["Edad"].max() - df["Edad"].min()
amplitud = math.ceil(rango / k)

# Paso 3: Crear los intervalos
limite_inferior = df["Edad"].min()
bins = [limite_inferior + i * amplitud for i in range(k + 1)]
frecuencias = pd.cut(df["Edad"], bins=bins, right=False)
tabla = frecuencias.value_counts().sort_index().reset_index()
tabla.columns = ["Intervalo", "Frecuencia"]

# Paso 4: Agregar marca de clase
tabla["Marca de clase"] = [(interval.left + interval.right) / 2 for interval in tabla["Intervalo"]]
tabla["FA"] = tabla["Frecuencia"].cumsum()

# Media agrupada
media = np.average(tabla["Marca de clase"], weights=tabla["Frecuencia"])

# Mediana agrupada
N = n
N_2 = N / 2
tabla["Frec. Acum. Ant."] = tabla["FA"] - tabla["Frecuencia"]

for i, row in tabla.iterrows():
    if row["FA"] >= N_2:
        LI = row["Intervalo"].left
        Fi = row["Frecuencia"]
        F_ant = row["Frec. Acum. Ant."]
        mediana = LI + ((N_2 - F_ant) / Fi) * amplitud
        break

# Moda agrupada
moda_idx = tabla["Frecuencia"].idxmax()
LI_moda = tabla.loc[moda_idx, "Intervalo"].left
f1 = tabla.loc[moda_idx, "Frecuencia"]
f0 = tabla.loc[moda_idx - 1, "Frecuencia"] if moda_idx > 0 else 0
f2 = tabla.loc[moda_idx + 1, "Frecuencia"] if moda_idx + 1 < len(tabla) else 0

moda = LI_moda + ((f1 - f0) / ((2 * f1) - f0 - f2)) * amplitud

# Mostrar resultados
print(f"Media agrupada: {media:.2f}")
print(f"Mediana agrupada: {mediana:.2f}")
print(f"Moda agrupada: {moda:.2f}")
```

# **ùüó. ùêáùêöùê•ùê•ùêû ùê¶ùêûùêùùê¢ùêùùêöùê¨ ùêùùêû ùê©ùê®ùê¨ùê¢ùêúùê¢√≥ùêß ùê©ùêöùê´ùêö ùêùùêöùê≠ùê®ùê¨ ùêöùê†ùê´ùêÆùê©ùêöùêùùê®ùê¨: ùêêùüè, ùêêùüë, ùêÉùüë, ùêÉùüñ ùêû ùê¢ùêßùê≠ùêûùê´ùê©ùê´ùêûùê≠ùêû ùê•ùê®ùê¨ ùêØùêöùê•ùê®ùê´ùêûùê¨ ùêûùêßùêúùê®ùêßùê≠ùê´ùêöùêùùê®ùê¨**

```python
import pandas as pd
import numpy as np
import math

# Usaremos la lista de edades ya indicada previamente
df = pd.DataFrame(edades, columns=["Edad"])

# Total de datos
n = len(df)

# N√∫mero de clases con regla de Sturges
k = round(1 + 3.322 * math.log10(n))

# Rango y amplitud
rango = df["Edad"].max() - df["Edad"].min()
amplitud = math.ceil(rango / k)

# Crear los intervalos (bins)
limite_inferior = df["Edad"].min()
bins = [limite_inferior + i * amplitud for i in range(k + 1)]

# Tabla de frecuencias
frecuencias = pd.cut(df["Edad"], bins=bins, right=False)
tabla = frecuencias.value_counts().sort_index().reset_index()
tabla.columns = ["Intervalo", "Frecuencia"]
tabla["Marca de clase"] = [(interval.left + interval.right) / 2 for interval in tabla["Intervalo"]]
tabla["Frec. Acumulada"] = tabla["Frecuencia"].cumsum()

# Funci√≥n para calcular medidas de posici√≥n agrupadas
def medida_posicion(porcentaje):
    pos = (porcentaje / 100) * n
    for i in range(len(tabla)):
        if tabla.loc[i, "Frec. Acumulada"] >= pos:
            Li = tabla.loc[i, "Intervalo"].left
            Fi = tabla.loc[i, "Frecuencia"]
            Fa = tabla.loc[i - 1, "Frec. Acumulada"] if i > 0 else 0
            return round(Li + ((pos - Fa) / Fi) * amplitud, 2)

# Calcular Q1, Q3, D3, D8
Q1 = medida_posicion(25)
Q3 = medida_posicion(75)
D3 = medida_posicion(30)
D8 = medida_posicion(80)

# Mostrar resultados
print("Medidas de posici√≥n para datos agrupados:")
print(f"Q1 (25%): {Q1}")
print(f"Q3 (75%): {Q3}")
print(f"D3 (30%): {D3}")
print(f"D8 (80%): {D8}")
```

**a) Q1 (Primer Cuartil) = 18.2**

El 25% de los estudiantes tienen menos de 18.2 a√±os, lo que indica que una cuarta parte del grupo pertenece al segmento m√°s joven de la poblaci√≥n.

**b) Q3 (Tercer Cuartil) = 19.45**

El 75% de los estudiantes tienen menos de 19.45 a√±os, y solo el 25% restante supera esa edad. Esto nos dice que la gran mayor√≠a se encuentra por debajo de los 19.5 a√±os.

**c) D3 (Tercer Decil) = 18.3**

El 30% de los estudiantes tienen menos de 18.3 a√±os, reforzando que hay una concentraci√≥n significativa en los primeros rangos de edad.

**d) D8 (Octavo Decil) = 19.68**

El 80% de los estudiantes tienen menos de 19.68 a√±os, por lo que solo el 20% de la muestra supera ese punto. El grupo de mayor edad representa una peque√±a fracci√≥n del total.

# **ùüèùüé. ùêÇùê®ùêßùê¨ùê≠ùê´ùêÆùê≤ùêö ùêÆùêß ùê¢ùêßùê≠ùêûùê´ùêØùêöùê•ùê® ùêùùêû ùêÇùê®ùêßùêüùê¢ùêöùêßùê≥ùêö ùêùùêû ùüóùüì% ùê©ùêöùê´ùêö ùê•ùêö ùê¶ùêûùêùùê¢ùêö ùê´ùêûùêöùê• ùêùùêû ùê•ùêö ùêßùê®ùê≠ùêö ùêùùêûùêüùê¢ùêßùê¢ùê≠ùê¢ùêØùêö ùêû ùê¢ùêßùê≠ùêûùê´ùê©ùê´ùêûùê≠ùêû ùêûùê• ùê´ùêûùê¨ùêÆùê•ùê≠ùêöùêùùê®.**

```python
import pandas as pd
import numpy as np
from scipy.stats import norm

# Simular 100 notas con distribuci√≥n realista
np.random.seed(42)  # Fijar semilla para reproducibilidad
notas = np.random.normal(loc=3.8, scale=0.4, size=100)
notas = np.clip(notas, 2.5, 5.0)  # Limitar valores entre 2.5 y 5.0

# Crear DataFrame
df = pd.DataFrame(notas, columns=["Nota"])

# Calcular estad√≠sticos
n = len(df)
media = df["Nota"].mean()
desviacion = df["Nota"].std(ddof=1)
z = norm.ppf(0.975)  # Valor z para 95% de confianza (1.96)

# Calcular intervalo de confianza
margen_error = z * (desviacion / np.sqrt(n))
IC_inf = media - margen_error
IC_sup = media + margen_error

# Resultados
print(f"Media muestral: {media:.2f}")
print(f"Intervalo de confianza del 95%: ({IC_inf:.2f}, {IC_sup:.2f})")
```

Con un nivel de confianza del 95%, podemos afirmar que la media real de las notas definitivas de la poblaci√≥n estudiantil se encuentra entre 3.74 y 3.90.
