```python
# @title **Ejemplo 1. Construye la gráfica de distribución normal con media 10 y desviación 2**

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define la media y la desviación estándar
media = 10
desviacion_estandar = 2

# Genera un rango de valores x para la gráfica
x = np.linspace(media - 4*desviacion_estandar, media + 4*desviacion_estandar, 100)

# Calcula la función de densidad de probabilidad (PDF) para cada valor de x
pdf = norm.pdf(x, media, desviacion_estandar)

# Crea la gráfica
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label=f'Media = {media}, Desviación estándar = {desviacion_estandar}')
plt.title('Gráfica de Distribución Normal - Santiago López Arcila')
plt.xlabel('Valores')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# @title **Ejemplo 5. Tres curvas con distintos colores y porcentajes**
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de la distribución
mu = 21
sigma = 3
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = norm.pdf(x, mu, sigma)

# Crear figura
fig, ax = plt.subplots(figsize=(10, 6))

# Dibujar las 3 curvas con distintos colores (aunque sean iguales)
ax.plot(x, y, label='Curva 1', color='blue')
ax.plot(x, y, label='Curva 2', color='orange')
ax.plot(x, y, label='Curva 3', color='green')

# Sombras para 1, 2 y 3 desviaciones estándar
# 1 SD
x1 = np.linspace(mu - sigma, mu + sigma, 1000)
ax.fill_between(x1, norm.pdf(x1, mu, sigma), color='blue', alpha=0.3, label='1 SD (68.27%)')

# 2 SD
x2 = np.linspace(mu - 2*sigma, mu + 2*sigma, 1000)
ax.fill_between(x2, norm.pdf(x2, mu, sigma), color='green', alpha=0.3, label='2 SD (95.45%)')

# 3 SD
x3 = np.linspace(mu - 3*sigma, mu + 3*sigma, 1000)
ax.fill_between(x3, norm.pdf(x3, mu, sigma), color='yellow', alpha=0.3, label='3 SD (99.73%)')

# Título y etiquetas
ax.set_title("Tres Curvas Normales con Media 21 y Desviación Estándar 3 - Santiago López Arcila")
ax.set_xlabel("X")
ax.set_ylabel("Probabilidad")
ax.legend(loc='center right')
ax.grid(True)

plt.show()
```

```python
# @title **Ejemplo 7. Curva normal estandar**
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de la curva normal estándar
mu = 0
sigma = 1

# Z que deja 96.78% del área a la izquierda
z_derecha = norm.ppf(0.9678)

# Rango de X y valores de la curva normal
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, mu, sigma)

# Área bajo la curva desde -∞ hasta z_derecha
x_fill = np.linspace(-4, z_derecha, 1000)
y_fill = norm.pdf(x_fill, mu, sigma)

# Gráfico
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, y, color='blue')
ax.fill_between(x_fill, y_fill, color='blue', alpha=0.4)

# Texto con el porcentaje en el área sombreada
ax.text(0, 0.1, "96.78%", fontsize=14, color='black', ha='center')

# Barra vertical solo en el extremo derecho
y_barra = norm.pdf(z_derecha)
ax.vlines(z_derecha, 0, y_barra, color='black', linewidth=4)
ax.hlines([y_barra * 0.9, y_barra * 0.1], z_derecha - 0.05, z_derecha + 0.05, color='black', linewidth=4)

# Ajustes del gráfico
ax.set_xlim(-4, 4)
ax.set_title("Curva Normal Estándar - Santiago López Arcila")
ax.set_xlabel("Valor Z")
ax.set_ylabel("Probabilidad")
ax.grid(True)

plt.show()
```

```python
# @title **Ejercicio 1. Resuelve: El volumen de hojas que una oficina de abogados deposita en un juzgados de sus casos atendidos es una variable aleatoria normal para la cual μ = 1200  y σ = 250. Determine la probabilidad de que el volumen depositado sea menor a 1050 hojas y mayor a 980. Haz el gráfico, sombrea el area, y coloca el porcentaje en el gráfico**

import matplotlib.pyplot as plt
import numpy as np
# @title **Resolución del Problema: Volumen de Hojas en Juzgados - Sharick Nadjar**

# Parámetros de la distribución normal
mu = 1200
sigma = 250

# Valores límite para la probabilidad
x1 = 1050
x2 = 980

# Calcular la probabilidad de que el volumen sea menor a 1050 hojas (P(X < 1050))
prob_menor_1050 = norm.cdf(x1, mu, sigma)
print(f"La probabilidad de que el volumen sea menor a 1050 hojas es: {prob_menor_1050:.4f}")

# Calcular la probabilidad de que el volumen sea mayor a 980 hojas (P(X > 980))
# P(X > 980) = 1 - P(X <= 980)
prob_mayor_980 = 1 - norm.cdf(x2, mu, sigma)
print(f"La probabilidad de que el volumen sea mayor a 980 hojas es: {prob_mayor_980:.4f}")

# Calcular la probabilidad de que el volumen esté entre 980 y 1050 hojas (P(980 < X < 1050))
# P(980 < X < 1050) = P(X < 1050) - P(X < 980)
prob_entre_980_1050 = norm.cdf(x1, mu, sigma) - norm.cdf(x2, mu, sigma)
print(f"La probabilidad de que el volumen esté entre 980 y 1050 hojas es: {prob_entre_980_1050:.4f}")

# Convertir la probabilidad a porcentaje
porcentaje_entre_980_1050 = prob_entre_980_1050 * 100
print(f"El porcentaje de probabilidad de que el volumen esté entre 980 y 1050 hojas es: {porcentaje_entre_980_1050:.2f}%")


# Generar rango de valores x para la gráfica
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# Calcular la función de densidad de probabilidad (PDF) para cada valor de x
pdf = norm.pdf(x, mu, sigma)

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label=f'μ = {mu}, σ = {sigma}')

# Sombrear el área entre 980 y 1050
x_fill = np.linspace(x2, x1, 100)
pdf_fill = norm.pdf(x_fill, mu, sigma)
plt.fill_between(x_fill, pdf_fill, color='skyblue', alpha=0.5, label=f'{porcentaje_entre_980_1050:.2f}% entre 980 y 1050')

# Añadir el porcentaje en el área sombreada
plt.text((x1 + x2) / 2, norm.pdf((x1 + x2) / 2, mu, sigma) * 0.5, f'{porcentaje_entre_980_1050:.2f}%', ha='center', fontsize=12, color='black')

# Configurar la gráfica
plt.title('Distribución del Volumen de Hojas Depositadas - Santiago López Arcila')
plt.xlabel('Volumen de Hojas')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# prompt: Sabemos que los paquetes de 1 libra de carne para hamburguesas tienen un contenido medio de 1.00 lb. con una desviación estándar de 0.13, Si tomamos al azar 50 paquetes de 1 libra de carne para hamburguesas calcule la probabilidad de que el peso medio de los paquetes sea mayor a 1.05 libras
# Resuelve como el ejercicio 1, acomoda los intervalos del eje x para que se pueda observar el porcentaje

import matplotlib.pyplot as plt
import numpy as np
# @title **Ejercicio 2. Resuelve: Sabemos que los paquetes de 1 libra de carne para hamburguesas tienen un contenido medio de 1.00 lb. con una desviación estándar de 0.13, Si tomamos al azar 50 paquetes de 1 libra de carne para hamburguesas calcule la probabilidad de que el peso medio de los paquetes sea mayor a 1.05 libras. Haz el gráfico, sombrea el area, y coloca el porcentaje en el gráfico**

# Parámetros de la distribución de la media muestral
media_poblacion = 1.00
desviacion_estandar_poblacion = 0.13
n = 50  # Tamaño de la muestra

# Calcular la desviación estándar de la media muestral (error estándar)
error_estandar = desviacion_estandar_poblacion / np.sqrt(n)

# Valor para el cual queremos calcular la probabilidad
x_barra = 1.05

# Calcular la probabilidad de que el peso medio sea mayor a 1.05 libras (P(X_barra > 1.05))
# P(X_barra > 1.05) = 1 - P(X_barra <= 1.05)
prob_mayor_1_05 = 1 - norm.cdf(x_barra, media_poblacion, error_estandar)
print(f"La probabilidad de que el peso medio de los paquetes sea mayor a {x_barra} libras es: {prob_mayor_1_05:.4f}")

# Convertir la probabilidad a porcentaje
porcentaje_mayor_1_05 = prob_mayor_1_05 * 100
print(f"El porcentaje de probabilidad de que el peso medio de los paquetes sea mayor a {x_barra} libras es: {porcentaje_mayor_1_05:.2f}%")

# Generar rango de valores x para la gráfica de la distribución de la media muestral
# Ajustar el rango para que se observe mejor el porcentaje
x = np.linspace(media_poblacion - 4 * error_estandar, media_poblacion + 4 * error_estandar, 1000)

# Calcular la función de densidad de probabilidad (PDF) para cada valor de x
pdf = norm.pdf(x, media_poblacion, error_estandar)

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label=f'Media = {media_poblacion:.2f}, Error Estándar = {error_estandar:.4f}')

# Sombrear el área mayor a 1.05
x_fill = np.linspace(x_barra, media_poblacion + 4 * error_estandar, 100)
pdf_fill = norm.pdf(x_fill, media_poblacion, error_estandar)
plt.fill_between(x_fill, pdf_fill, color='lightcoral', alpha=0.5, label=f'{porcentaje_mayor_1_05:.2f}% > {x_barra}')

# Añadir el porcentaje en el área sombreada
# Colocar el texto un poco a la derecha del valor límite
plt.text(x_barra + error_estandar * 0.5, norm.pdf(x_barra + error_estandar * 0.5, media_poblacion, error_estandar) * 0.5, f'{porcentaje_mayor_1_05:.2f}%', ha='left', fontsize=12, color='black')

# Configurar la gráfica
plt.title('Distribución Muestral del Peso Medio de los Paquetes de Carne - Santiago López Arcila')
plt.xlabel('Peso Medio de los Paquetes (lb)')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# prompt: Una máquina despachadora de nieve está ajustada para servir un promedio de 30 ml por cono. Si la cantidad de nieve está normalmente distribuida con una varianza de 225 ml. ¿Cuál es la probabilidad de de que un cono contenga entre 25 ml y 35 ml?
# Resuelve como el ejercicio 1

import matplotlib.pyplot as plt
import numpy as np
# @title **Ejercicio 3. Resuelve: Una máquina despachadora de nieve está ajustada para servir un promedio de 30 ml por cono. Si la cantidad de nieve está normalmente distribuida con una varianza de 225 ml. ¿Cuál es la probabilidad de de que un cono contenga entre 25 ml y 35 ml?**

# Parámetros de la distribución normal
media = 30
varianza = 225
desviacion_estandar = np.sqrt(varianza) # Calcular la desviación estándar a partir de la varianza

# Valores límite para la probabilidad
x1 = 25
x2 = 35

# Calcular la probabilidad de que un cono contenga menos de 35 ml (P(X < 35))
prob_menor_35 = norm.cdf(x2, media, desviacion_estandar)
print(f"La probabilidad de que un cono contenga menos de {x2} ml es: {prob_menor_35:.4f}")

# Calcular la probabilidad de que un cono contenga menos de 25 ml (P(X < 25))
prob_menor_25 = norm.cdf(x1, media, desviacion_estandar)
print(f"La probabilidad de que un cono contenga menos de {x1} ml es: {prob_menor_25:.4f}")

# Calcular la probabilidad de que un cono contenga entre 25 ml y 35 ml (P(25 < X < 35))
# P(25 < X < 35) = P(X < 35) - P(X < 25)
prob_entre_25_35 = prob_menor_35 - prob_menor_25
print(f"La probabilidad de que un cono contenga entre {x1} ml y {x2} ml es: {prob_entre_25_35:.4f}")

# Convertir la probabilidad a porcentaje
porcentaje_entre_25_35 = prob_entre_25_35 * 100
print(f"El porcentaje de probabilidad de que un cono contenga entre {x1} ml y {x2} ml es: {porcentaje_entre_25_35:.2f}%")

# Generar rango de valores x para la gráfica
x = np.linspace(media - 4*desviacion_estandar, media + 4*desviacion_estandar, 1000)

# Calcular la función de densidad de probabilidad (PDF) para cada valor de x
pdf = norm.pdf(x, media, desviacion_estandar)

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label=f'μ = {media}, σ = {desviacion_estandar}')

# Sombrear el área entre 25 y 35
x_fill = np.linspace(x1, x2, 100)
pdf_fill = norm.pdf(x_fill, media, desviacion_estandar)
plt.fill_between(x_fill, pdf_fill, color='skyblue', alpha=0.5, label=f'{porcentaje_entre_25_35:.2f}% entre {x1} y {x2}')

# Añadir el porcentaje en el área sombreada
plt.text((x1 + x2) / 2, norm.pdf((x1 + x2) / 2, media, desviacion_estandar) * 0.5, f'{porcentaje_entre_25_35:.2f}%', ha='center', fontsize=12, color='black')

# Configurar la gráfica
plt.title('Distribución de la Cantidad de Nieve Despachada - Santiago López Arcila')
plt.xlabel('Cantidad de Nieve (ml)')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# prompt: La cantidad real de café instantáneo que una máquina vierte en frascos de 4 onzas, puede considerarse como una distribución normal que tiene una media de 4,082 onzas y una desviación estándar de 0.04. ¿Porcentaje frascos se llenan con más de 4,090 onzas serán desechados por control de calidad?
# Resuelve como el ejercicio 1

import matplotlib.pyplot as plt
import numpy as np
# @title **Ejercicio 4. Resuelve: La cantidad real de café instantáneo que una máquina vierte en frascos de 4 onzas, puede considerarse como una distribución normal que tiene una media de 4,082 onzas y una desviación estándar de 0.04. ¿Porcentaje frascos se llenan con más de 4,090 onzas serán desechados por control de calidad?**

# Parámetros de la distribución normal
media = 4.082
desviacion_estandar = 0.04

# Valor límite para la probabilidad
x_desechar = 4.090

# Calcular la probabilidad de que un frasco se llene con más de 4,090 onzas (P(X > 4.090))
# P(X > 4.090) = 1 - P(X <= 4.090)
prob_mayor_4_090 = 1 - norm.cdf(x_desechar, media, desviacion_estandar)
print(f"La probabilidad de que un frasco se llene con más de {x_desechar} onzas es: {prob_mayor_4_090:.4f}")

# Convertir la probabilidad a porcentaje
porcentaje_desechado = prob_mayor_4_090 * 100
print(f"El porcentaje de frascos que serán desechados por control de calidad es: {porcentaje_desechado:.2f}%")

# Generar rango de valores x para la gráfica
x = np.linspace(media - 4*desviacion_estandar, media + 4*desviacion_estandar, 1000)

# Calcular la función de densidad de probabilidad (PDF) para cada valor de x
pdf = norm.pdf(x, media, desviacion_estandar)

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label=f'μ = {media}, σ = {desviacion_estandar}')

# Sombrear el área mayor a 4.090
x_fill = np.linspace(x_desechar, media + 4*desviacion_estandar, 100)
pdf_fill = norm.pdf(x_fill, media, desviacion_estandar)
plt.fill_between(x_fill, pdf_fill, color='salmon', alpha=0.5, label=f'{porcentaje_desechado:.2f}% > {x_desechar}')

# Añadir el porcentaje en el área sombreada
# Colocar el texto un poco a la derecha del valor límite
plt.text(x_desechar + desviacion_estandar * 0.5, norm.pdf(x_desechar + desviacion_estandar * 0.5, media, desviacion_estandar) * 0.5, f'{porcentaje_desechado:.2f}%', ha='left', fontsize=12, color='black')

# Configurar la gráfica
plt.title('Distribución de la Cantidad de Café en Frascos - Santiago López Arcila')
plt.xlabel('Cantidad de Café (onzas)')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# prompt: Un fabricante especifica que su producto sigue el comportamiento de una distribución normal con duración media de 4.5 años con una desviación estándar de 0.5 años. Cuál es la probabilidad de que al adquirir una unidad al azar tenga una duración ente 3.8 y 4.7 años?
# Resuelve como el ejercicio 1

import matplotlib.pyplot as plt
import numpy as np
# @title **Ejercicio 5. Resuelve: Duración de un producto**

# Parámetros de la distribución normal
media = 4.5
desviacion_estandar = 0.5

# Valores límite para la probabilidad
x1 = 3.8
x2 = 4.7

# Calcular la probabilidad de que la duración sea menor a 4.7 años (P(X < 4.7))
prob_menor_4_7 = norm.cdf(x2, media, desviacion_estandar)
print(f"La probabilidad de que la duración sea menor a {x2} años es: {prob_menor_4_7:.4f}")

# Calcular la probabilidad de que la duración sea menor a 3.8 años (P(X < 3.8))
prob_menor_3_8 = norm.cdf(x1, media, desviacion_estandar)
print(f"La probabilidad de que la duración sea menor a {x1} años es: {prob_menor_3_8:.4f}")

# Calcular la probabilidad de que la duración esté entre 3.8 y 4.7 años (P(3.8 < X < 4.7))
# P(3.8 < X < 4.7) = P(X < 4.7) - P(X < 3.8)
prob_entre_3_8_4_7 = prob_menor_4_7 - prob_menor_3_8
print(f"La probabilidad de que la duración esté entre {x1} y {x2} años es: {prob_entre_3_8_4_7:.4f}")

# Convertir la probabilidad a porcentaje
porcentaje_entre_3_8_4_7 = prob_entre_3_8_4_7 * 100
print(f"El porcentaje de probabilidad de que la duración esté entre {x1} y {x2} años es: {porcentaje_entre_3_8_4_7:.2f}%")

# Generar rango de valores x para la gráfica
x = np.linspace(media - 4*desviacion_estandar, media + 4*desviacion_estandar, 1000)

# Calcular la función de densidad de probabilidad (PDF) para cada valor de x
pdf = norm.pdf(x, media, desviacion_estandar)

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label=f'μ = {media}, σ = {desviacion_estandar}')

# Sombrear el área entre 3.8 y 4.7
x_fill = np.linspace(x1, x2, 100)
pdf_fill = norm.pdf(x_fill, media, desviacion_estandar)
plt.fill_between(x_fill, pdf_fill, color='skyblue', alpha=0.5, label=f'{porcentaje_entre_3_8_4_7:.2f}% entre {x1} y {x2}')

# Añadir el porcentaje en el área sombreada
plt.text((x1 + x2) / 2, norm.pdf((x1 + x2) / 2, media, desviacion_estandar) * 0.5, f'{porcentaje_entre_3_8_4_7:.2f}%', ha='center', fontsize=12, color='black')

# Configurar la gráfica
plt.title('Distribución de la Duración del Producto - Santiago López Arcila')
plt.xlabel('Duración (años)')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# prompt: Se sabe que la prueba de auto estima sigue una distribución normal con una media de 60, con una desviación estándar de 5.83. Si se aplica una prueba de autoestima a 25 personas quienes obtienen una calificación promedio de 62.1 ¿Cuál es la probabilidad de que la media muestral sea no mas que 62.1?
# Resuelve como el ejercicio 1

import matplotlib.pyplot as plt
import numpy as np
# @title **Ejercicio 6. Resuelve: Probabilidad de la media muestral de autoestima - Sharick Nadjar**

# Parámetros de la distribución poblacional
media_poblacion_autoestima = 60
desviacion_estandar_poblacion_autoestima = 5.83
n_muestra_autoestima = 25 # Tamaño de la muestra

# Calcular la desviación estándar de la media muestral (error estándar) para la autoestima
error_estandar_autoestima = desviacion_estandar_poblacion_autoestima / np.sqrt(n_muestra_autoestima)

# Valor de la media muestral observada
media_muestral_observada = 62.1

# Calcular la probabilidad de que la media muestral sea no más que 62.1 (P(X_barra <= 62.1))
prob_no_mas_62_1 = norm.cdf(media_muestral_observada, media_poblacion_autoestima, error_estandar_autoestima)
print(f"La probabilidad de que la media muestral de autoestima sea no más que {media_muestral_observada} es: {prob_no_mas_62_1:.4f}")

# Convertir la probabilidad a porcentaje
porcentaje_no_mas_62_1 = prob_no_mas_62_1 * 100
print(f"El porcentaje de probabilidad de que la media muestral de autoestima sea no más que {media_muestral_observada} es: {porcentaje_no_mas_62_1:.2f}%")

# Generar rango de valores x para la gráfica de la distribución de la media muestral
x = np.linspace(media_poblacion_autoestima - 4 * error_estandar_autoestima, media_poblacion_autoestima + 4 * error_estandar_autoestima, 1000)

# Calcular la función de densidad de probabilidad (PDF) para cada valor de x
pdf = norm.pdf(x, media_poblacion_autoestima, error_estandar_autoestima)

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label=f'Media = {media_poblacion_autoestima:.2f}, Error Estándar = {error_estandar_autoestima:.4f}')

# Sombrear el área a la izquierda de 62.1
x_fill = np.linspace(media_poblacion_autoestima - 4 * error_estandar_autoestima, media_muestral_observada, 1000)
pdf_fill = norm.pdf(x_fill, media_poblacion_autoestima, error_estandar_autoestima)
plt.fill_between(x_fill, pdf_fill, color='skyblue', alpha=0.5, label=f'{porcentaje_no_mas_62_1:.2f}% <= {media_muestral_observada}')

# Añadir el porcentaje en el área sombreada
# Colocar el texto un poco a la izquierda del valor límite
plt.text(media_muestral_observada - error_estandar_autoestima * 0.5, norm.pdf(media_muestral_observada - error_estandar_autoestima * 0.5, media_poblacion_autoestima, error_estandar_autoestima) * 0.5, f'{porcentaje_no_mas_62_1:.2f}%', ha='right', fontsize=12, color='black')

# Configurar la gráfica
plt.title('Distribución Muestral de la Media de Autoestima - Santiago López Arcila')
plt.xlabel('Media Muestral de Autoestima')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)
plt.legend()
plt.show()
```
