```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import math

# Datos proporcionados
edades = np.array([
    19, 18, 18, 21, 18, 18, 18, 17, 18, 22, 18, 18, 18, 18, 18, 22, 21, 19, 18, 19,
    18, 19, 18, 17, 19, 20, 19, 18, 17, 18, 20, 19, 20, 22, 18, 19, 19, 18, 19, 20,
    19, 18, 17, 19, 19, 19, 19, 18, 17, 18, 19, 20, 23, 18, 18, 20, 18, 18, 18, 18,
    18, 19, 17, 17, 20, 19, 18, 19, 19, 21, 19, 21, 17, 19, 18, 18, 17, 17, 18, 18,
    17, 18, 19, 17, 18, 18, 18, 18, 20, 18, 18, 18, 17, 19, 18, 17, 18, 18, 18, 18,
    18, 18, 18, 18, 19, 18, 19, 18, 18, 17, 18, 18, 18, 18, 18, 18, 18, 18, 24, 18,
    19, 19, 18, 17, 17, 17, 20, 17, 18, 17, 18, 18, 19, 18, 18, 18
])
```

# --- Punto 1: Medidas de Tendencia Central y Dispersión ---

```python
media = np.mean(edades)
varianza = np.var(edades, ddof=1)  # ddof=1 para varianza muestral
desviacion = np.std(edades, ddof=1)

print("1. Medidas de Tendencia Central y Dispersión:")
print(f"  - Media: {media:.2f} años")
print(f"  - Varianza: {varianza:.2f} años²")
print(f"  - Desviación Estándar: {desviacion:.2f} años\n")
```

# --- Punto 2: Medidas de Posición (Cuartiles, Deciles) ---

```python
Q1 = np.percentile(edades, 25)
Q3 = np.percentile(edades, 75)
D3 = np.percentile(edades, 30)
D8 = np.percentile(edades, 80)

print("2. Medidas de Posición:")
print(f"  - Q1 (Primer Cuartil): {Q1:.2f} años (25% de los datos ≤ {Q1:.2f})")
print(f"  - Q3 (Tercer Cuartil): {Q3:.2f} años (75% de los datos ≤ {Q3:.2f})")
print(f"  - D3 (Tercer Decil): {D3:.2f} años (30% de los datos ≤ {D3:.2f})")
print(f"  - D8 (Octavo Decil): {D8:.2f} años (80% de los datos ≤ {D8:.2f})\n")
```

# --- Punto 3: Diagrama de Caja y Bigotes ---

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=edades, color='skyblue')
plt.title("3. Diagrama de Caja y Bigotes de EDAD - Josue Ortega")
plt.xlabel("Edad (años)")
plt.show()
print("3. Diagrama de Caja: Se observan posibles valores atípicos alrededor de 23-24 años.\n")
```

# --- Punto 4: Histograma y Polígono de Frecuencias ---

```python
plt.figure(figsize=(10, 6))
freq, bins, _ = plt.hist(edades, bins=range(16, 26, 2), edgecolor='black', alpha=0.7, label='Frecuencia Absoluta')
plt.plot(bins[:-1] + 1, freq, 'ro-', label='Polígono de Frecuencia')  # Puntos medios
plt.title("4. Histograma y Polígono de Frecuencias de EDAD - Josué Ortega")
plt.xlabel("Edad (años)")
plt.ylabel("Frecuencia")
plt.legend()
plt.show()
```

# --- Punto 5: Intervalo de Confianza para la Media (95%) ---

```python
n = len(edades)
t_score = stats.t.ppf(0.975, df=n-1)  # Valor crítico t
error_estandar = desviacion / np.sqrt(n)
IC_media_inf = media - t_score * error_estandar
IC_media_sup = media + t_score * error_estandar

print("5. Intervalo de Confianza para la Media Poblacional (95%):")
print(f"  - IC: ({IC_media_inf:.2f}, {IC_media_sup:.2f}) años")
print("  - Interpretación: Con 95% de confianza, la media poblacional de EDAD está entre estos valores.\n")
```


# --- Punto 6: Intervalo de Confianza para la Varianza (95%) ---

```python
chi2_inf = stats.chi2.ppf(0.025, df=n-1)
chi2_sup = stats.chi2.ppf(0.975, df=n-1)
IC_var_inf = (n-1) * varianza / chi2_sup
IC_var_sup = (n-1) * varianza / chi2_inf

print("6. Intervalo de Confianza para la Varianza Poblacional (95%):")
print(f"  - IC: ({IC_var_inf:.2f}, {IC_var_sup:.2f}) años²")
print("  - Interpretación: Con 95% de confianza, la varianza poblacional de EDAD está entre estos valores.\n")
```

# --- Punto 7: Muestras (60% y 40%) ---

```python
np.random.seed(42)  # Para reproducibilidad
muestra1 = np.random.choice(edades, size=int(0.6 * n), replace=False)
muestra2 = np.array([x for x in edades if x not in muestra1])

# Medidas para Muestra 1
media_m1 = np.mean(muestra1)
var_m1 = np.var(muestra1, ddof=1)
std_m1 = np.std(muestra1, ddof=1)

# Medidas para Muestra 2
media_m2 = np.mean(muestra2)
var_m2 = np.var(muestra2, ddof=1)
std_m2 = np.std(muestra2, ddof=1)

print("7. Medidas para Muestras:")
print("  - Muestra 1 (60%):")
print(f"    Media: {media_m1:.2f} años, Varianza: {var_m1:.2f} años², Desviación: {std_m1:.2f} años")
print("  - Muestra 2 (40%):")
print(f"    Media: {media_m2:.2f} años, Varianza: {var_m2:.2f} años², Desviación: {std_m2:.2f} años\n")
```

# --- Punto 8: Diagrama de Caja para Ambas Muestras ---

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=[muestra1, muestra2], palette="Set2")
plt.xticks([0, 1], ["Muestra 1 (60%)", "Muestra 2 (40%)"])
plt.title("8. Comparación de Muestras - Josué Ortega")
plt.xlabel("Muestras")
plt.ylabel("Edad (años)")
plt.show()
```

# --- Punto 9: IC para Diferencia de Medias (95%) ---
# Asumiendo varianzas iguales (usamos prueba t de Student)

```python
t_score_diff = stats.t.ppf(0.975, df=n-2)
pooled_var = ((len(muestra1)-1)*var_m1 + (len(muestra2)-1)*var_m2) / (n-2)
se_diff = np.sqrt(pooled_var * (1/len(muestra1) + 1/len(muestra2)))
diff_media = media_m1 - media_m2

IC_diff_inf = diff_media - t_score_diff * se_diff
IC_diff_sup = diff_media + t_score_diff * se_diff

print("9. Intervalo de Confianza para Diferencia de Medias (95%):")
print(f"  - IC: ({IC_diff_inf:.2f}, {IC_diff_sup:.2f}) años")
print("  - Interpretación: Si el IC incluye el 0, no hay evidencia de diferencia significativa entre las medias.\n")
```

# --- Punto 10: IC para Cociente de Varianzas (95%) ---

```python
F_inf = stats.f.ppf(0.025, dfn=len(muestra1)-1, dfd=len(muestra2)-1)
F_sup = stats.f.ppf(0.975, dfn=len(muestra1)-1, dfd=len(muestra2)-1)
IC_ratio_var_inf = (var_m1 / var_m2) / F_sup
IC_ratio_var_sup = (var_m1 / var_m2) / F_inf

print("10. Intervalo de Confianza para Cociente de Varianzas (95%):")
print(f"  - IC: ({IC_ratio_var_inf:.2f}, {IC_ratio_var_sup:.2f})")
print("  - Interpretación: Si el IC incluye el 1, no hay evidencia de diferencia significativa entre las varianzas.")
```
