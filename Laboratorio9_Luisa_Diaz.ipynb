```python
# prompt: genera un codigo en el que tambien aparezca la distribucion t-student en el intervalo de confianza

# @title **e. distribucion t-student - Descriptores numéricos Bivariados 'DEFINITIVA vs ESTRATO'**
print( 'DEFINITIVA vs ESTRATO')
df2 = df.groupby ('ESTRATO') ['DEFINITIVA']. agg(['mean', 'count', 'var', 'std'])
df2[' coef_var(%) '] = round((df2['std'] / df2['mean']) * 100,1)
df2['std_mean'] = round(df2['std'] / np.sqrt(df2['count']),3)

# Degrees of freedom
df2['df'] = df2['count'] - 1

# Calculate t-value for 95% confidence interval
alpha = 0.05
df2['t_value'] = df2['df'].apply(lambda df: stats.t.ppf(1 - alpha/2, df))

# Calculate the margin of error using the t-distribution
df2['margin_of_error'] = df2['t_value'] * df2['std_mean']

# Calculate the confidence interval using the t-distribution
df2['IC(µ) - 95% (t)'] = df2.apply(lambda row: (round(row['mean'] - row['margin_of_error'], 2), round(row['mean'] + row['margin_of_error'], 2)), axis=1)

df2
```

```python
# prompt: grafico de medias para los intervalos de confianza

import matplotlib.pyplot as plt

# Assuming df2 is the DataFrame from the previous code block
# with 'ESTRATO', 'mean', and 'IC(µ) - 95% (t)' columns

# Extract data for plotting
estratos = df2.index
means = df2['mean']
intervals = df2['IC(µ) - 95% (t)']

# Create the plot
plt.figure(figsize=(10, 6))
plt.errorbar(estratos, means, yerr=[(mean - lower) for mean, (lower, upper) in zip(means, intervals)],
             fmt='o', capsize=5, label='95% Confidence Interval')
plt.xlabel('Estrato')
plt.ylabel('Media de la Definitiva')
plt.title('f. Intervalos de Confianza de la Media de la Definitiva por Estrato - Luisa Diaz')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
