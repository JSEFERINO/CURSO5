```python
# @title **Ejemplo 1: construye la grafica de la distribuccion normal con media 10 y desviacion 2**

import numpy as np
import matplotlib.pyplot as plt

# Parmetros de la distribucin normal
mu = 10  # Media
sigma = 2  # Desviacin estndar

# Generar datos para la grfica
x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 100)  # Rango de valores en la x
y = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma)**2)  # Funcin de densidad de probabilidad

# Crear la grfica
plt.plot(x, y)

# Agregar etiquetas y ttulo
plt.xlabel('Valores')
plt.ylabel('Probabilidad')
plt.title('Distribucin Normal (Media = 10, Desviacin Estndar = 2)')

# Agregar cuadrcula
plt.grid(True)

# Mostrar la grfica
plt.show()
```

```python
# @title **Ejemplo 2: construye un grafico donde me dibujes las curvas normales con media 10 y diferentes desviaciones**

import numpy as np
import matplotlib.pyplot as plt

# Parmetros fijos
mu = 10

# Diferentes desviaciones estndar
sigmas = [1, 2, 3, 4]

# Generar datos y graficar para cada desviacin estndar
for sigma in sigmas:
    x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 100)
    y = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma)**2)
    plt.plot(x, y, label=f'Desviacin Estndar = {sigma}')

# Agregar etiquetas y ttulo
plt.xlabel('Valores')
plt.ylabel('Probabilidad')
plt.title('Distribucin Normal con Diferentes Desviaciones Estndar (Media = 10)')

# Agregar leyenda
plt.legend()

# Agregar cuadrcula
plt.grid(True)

# Mostrar la grfica
plt.show()
```

```python
# @title **Ejemplo 3: construye un grafico donde me dibujes diferentes curvas normales con media 8,10,12 y 15 y una desviacion de 4**

import numpy as np
import matplotlib.pyplot as plt

# Parmetros fijos
sigma = 4

# Diferentes medias
mus = [8, 10, 12, 15]

# Rango de valores en x (ajustado para abarcar todas las distribuciones)
x = np.linspace(min(mus) - 4 * sigma, max(mus) + 4 * sigma, 500)

# Crear la grfica para cada media
for mu in mus:
    y = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma)**2)
    plt.plot(x, y, label=f'Media = {mu}')

# Agregar etiquetas y ttulo
plt.xlabel('Valores')
plt.ylabel('Probabilidad')
plt.title('Distribucin Normal (Desviacin Estndar = 4, Varias Medias)')
plt.legend()
plt.grid(True)

# Mostrar la grfica
plt.show()
```

```python
# @title **Ejemplo 4: Dibuja 3 curvas normales con media 21 y desviacion 3 y muestrame en ellas el porcentaje correspondiente a una desviacion estandar, a dos desviacion estandar y a tres desviacion estandar**

import numpy as np
import matplotlib.pyplot as plt

# Parmetros fijos
mu = 21
sigma = 3

# Rango de valores en x
x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 500)

# Porcentajes de la regla emprica
percentages = {
    1: 68.27,
    2: 95.45,
    3: 99.73
}

# Crear la grfica para cada desviacin estndar
for s in range(1, 4):
    y = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma)**2)
    plt.plot(x, y, label=f'{s} Desviacion Estandar')

    # Calcular los puntos de desviacin estndar
    x1 = mu - s * sigma
    x2 = mu + s * sigma

    # Agregar sombreado para la regin de la desviacin estndar
    plt.fill_between(x, 0, y, where=((x > x1) & (x < x2)), alpha=0.3)
    plt.text((x1 + x2) / 2, max(y) / 2, f'{percentages[s]}%', ha='center', va='center', color='black')


# Agregar etiquetas y ttulo
plt.xlabel('Valores')
plt.ylabel('Probabilidad')
plt.title('Distribucin Normal (Media = 21, Desviacin Estndar = 3)')
plt.legend()
plt.grid(True)

# Mostrar la grfica
plt.show()
```

```python
# @title **Ejemplo 5: dejame los colores como estan y en cada region sombreada muestrame el porcentaje correspondiente**

import numpy as np
import matplotlib.pyplot as plt

# Parmetros fijos
mu = 21
sigma = 3

# Rango de valores en x
x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 500)

# Porcentajes de la regla emprica
percentages = {
    1: 68.27,
    2: 95.45,
    3: 99.73
}

# Colores para las regiones sombreadas (manteniendo los colores originales)
colors = ['skyblue', 'lightcoral', 'lightgreen']

# Crear la grfica para cada desviacin estndar
for i, s in enumerate(range(1, 4)):
    y = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma)**2)
    plt.plot(x, y, label=f'{s} Desviacion Estandar')

    # Calcular los puntos de desviacin estndar
    x1 = mu - s * sigma
    x2 = mu + s * sigma

    # Agregar sombreado para la regin de la desviacin estndar con el color correspondiente
    plt.fill_between(x, 0, y, where=((x > x1) & (x < x2)), alpha=0.3, color=colors[i])

    # Mostrar el porcentaje en la regin sombreada
    plt.text((x1 + x2) / 2, max(y) / 2, f'{percentages[s]}%', ha='center', va='center', color='black')


# Agregar etiquetas y ttulo
plt.xlabel('Valores')
plt.ylabel('Probabilidad')
plt.title('Distribucin Normal (Media = 21, Desviacin Estndar = 3)')
plt.legend()
plt.grid(True)

# Mostrar la grfica
plt.show()
```

```python
# @title **Ejemplo 7: Usa lo anterior y me hallas el area a la izquierda de 1.85. Dibuja la region y me escribes en el grafico el porcentaje**

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parmetros de la distribucin normal estndar
mu = 0
sigma = 1

# Valor para el cual se quiere calcular el rea a la izquierda
x = 1.85

# Generar datos para la grfica
x_values = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 500)
y_values = norm.pdf(x_values, loc=mu, scale=sigma)

# Calcular el rea a la izquierda de x
area = norm.cdf(x, loc=mu, scale=sigma)

# Crear la grfica
plt.plot(x_values, y_values)

# Sombrear el rea a la izquierda de x
x_fill = np.linspace(mu - 4 * sigma, x, 500)
y_fill = norm.pdf(x_fill, loc=mu, scale=sigma)
plt.fill_between(x_fill, y_fill, color='skyblue', alpha=0.5)

# Agregar una lnea vertical en x
plt.axvline(x, color='red', linestyle='--')

# Agregar texto con el porcentaje
plt.text(x - 0.5, 0.1, f'{area * 100:.2f}%', fontsize=12, color='red')

# Agregar etiquetas y ttulo
plt.xlabel('Valores')
plt.ylabel('Probabilidad')
plt.title(f'Distribucin Normal Estndar (rea a la Izquierda de {x})')
plt.grid(True)

# Mostrar la grfica
plt.show()
```

```python
# prompt: resuelve: El volumen de hojas que una oficina de abogados deposita en un juzgados de sus casos atendidos es una variable aleatoria normal para la cual  = 1200  y  = 250. Determine la probabilidad de que el volumen depositado sea menor a 1050 hojas y mayor a 980 y coloca el porcentaje en el grafico y pon el problema como @title

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# @title **El volumen de hojas que una oficina de abogados deposita en un juzgado**

# Parmetros de la distribucin normal
mu = 1200  # Media
sigma = 250  # Desviacin estndar

# Valores para calcular la probabilidad
x1 = 1050
x2 = 980

# Calcular la probabilidad de que el volumen sea menor a 1050
prob_menor_1050 = norm.cdf(x1, loc=mu, scale=sigma)

# Calcular la probabilidad de que el volumen sea mayor a 980
prob_mayor_980 = 1 - norm.cdf(x2, loc=mu, scale=sigma)


# Generar datos para la grfica
x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 500)
y = norm.pdf(x, loc=mu, scale=sigma)

# Crear la grfica
plt.plot(x, y)

# Sombrear el rea para x < 1050
x_fill1 = np.linspace(mu - 4 * sigma, x1, 500)
y_fill1 = norm.pdf(x_fill1, loc=mu, scale=sigma)
plt.fill_between(x_fill1, y_fill1, color='skyblue', alpha=0.5)
plt.text(x1 - 200, 0.0005, f'{prob_menor_1050*100:.2f}%', fontsize=10, color='blue')


# Sombrear el rea para x > 980
x_fill2 = np.linspace(x2, mu + 4 * sigma, 500)
y_fill2 = norm.pdf(x_fill2, loc=mu, scale=sigma)
plt.fill_between(x_fill2, y_fill2, color='lightcoral', alpha=0.5)
plt.text(x2 + 100, 0.0005, f'{prob_mayor_980*100:.2f}%', fontsize=10, color='red')

# Agregar lneas verticales
plt.axvline(x1, color='blue', linestyle='--')
plt.axvline(x2, color='red', linestyle='--')


# Agregar etiquetas y ttulo
plt.xlabel('Volumen de hojas')
plt.ylabel('Probabilidad')
plt.title('Distribucin del volumen de hojas depositadas')
plt.grid(True)

# Mostrar la grfica
plt.show()
```

```python
# @title **Sabemos que los paquetes de 1 libra de carne para hamburguesas tienen un contenido medio de 1.00 lb. con una desviacin estndar de 0.13, Si tomamos al azar 50 paquetes de 1 libra de carne para hamburguesas calcule la probabilidad de que el peso medio de los paquetes sea mayor a 1.05 libras**

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parmetros de la distribucin
mu = 1.00  # Media
sigma = 0.13  # Desviacin estndar
n = 50  # Tamao de la muestra
x_bar = 1.05  # Peso medio a evaluar

# Distribucin muestral de la media
sigma_x_bar = sigma / np.sqrt(n)

# Calcular la probabilidad de que la media muestral sea mayor a 1.05
z = (x_bar - mu) / sigma_x_bar
probabilidad = 1 - norm.cdf(z)

# Generar datos para la grfica
x = np.linspace(mu - 4 * sigma_x_bar, mu + 4 * sigma_x_bar, 500)
y = norm.pdf(x, loc=mu, scale=sigma_x_bar)

# Crear la grfica
plt.plot(x, y)

# Sombrear el rea para x_bar > 1.05
x_fill = np.linspace(x_bar, mu + 4 * sigma_x_bar, 500)
y_fill = norm.pdf(x_fill, loc=mu, scale=sigma_x_bar)
plt.fill_between(x_fill, y_fill, color='skyblue', alpha=0.5)

# Agregar lnea vertical en x_bar
plt.axvline(x_bar, color='red', linestyle='--')

# Agregar texto con la probabilidad
plt.text(x_bar + 0.01, 0.5, f'{probabilidad * 100:.2f}%', fontsize=12, color='red')


# Agregar etiquetas y ttulo
plt.xlabel('Peso medio de los paquetes (libras)')
plt.ylabel('Probabilidad')
plt.title('Sabemos que los paquetes de 1 libra de carne para hamburguesas tienen un contenido medio de 1.00 lb. con una desviacin estndar de 0.13, Si tomamos al azar 50 paquetes de 1 libra de carne para hamburguesas calcule la probabilidad de que el peso medio de los paquetes sea mayor a 1.05 libras')
plt.grid(True)

# Mostrar la grfica
plt.show()
```

```python
# prompt: resuelve igual que el anterior con la grafica: Un fabricante especifica que su producto sigue el comportamiento de una distribucin normal con duracin media de 4.5 aos con una desviacin estndar de 0.5 aos. Cul es la probabilidad de que al adquirir una unidad al azar tenga una duracin ente 3.8 y 4.7 aos?

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parmetros de la distribucin normal
mu = 4.5  # Media
sigma = 0.5  # Desviacin estndar

# Rango de valores en x
x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 500)

# Funcin de densidad de probabilidad
y = norm.pdf(x, loc=mu, scale=sigma)

# Calcular la probabilidad entre 3.8 y 4.7 aos
prob_38_47 = norm.cdf(4.7, loc=mu, scale=sigma) - norm.cdf(3.8, loc=mu, scale=sigma)


# Crear la grfica
plt.plot(x, y)

# Sombrear el rea entre 3.8 y 4.7
x_fill = np.linspace(3.8, 4.7, 500)
y_fill = norm.pdf(x_fill, loc=mu, scale=sigma)
plt.fill_between(x_fill, y_fill, color='skyblue', alpha=0.5)

# Agregar lneas verticales en los lmites
plt.axvline(3.8, color='red', linestyle='--')
plt.axvline(4.7, color='red', linestyle='--')

# Agregar texto con la probabilidad
plt.text((3.8 + 4.7) / 2, max(y) / 2, f'{prob_38_47 * 100:.2f}%', ha='center', va='center', color='black')


# Agregar etiquetas y ttulo
plt.xlabel('Duracin (aos)')
plt.ylabel('Probabilidad')
plt.title('Probabilidad de duracin entre 3.8 y 4.7 aos')
plt.grid(True)

# Mostrar la grfica
plt.show()

print(f"La probabilidad de que una unidad tenga una duracin entre 3.8 y 4.7 aos es: {prob_38_47 * 100:.2f}%")
```

```python
# prompt: resuelve igual que el anterior: Las Estaturas de los Estudiantes del prof. Hurtado siguen una distribucin Normal con Media de 170 cms y desviacin de 8.5. Si tomamos una muestra de 25 estudiantes, calcula la Probabilidad de que un estudiante seleccionado al azar tenga  Estatura promedio mayor a 175 cms.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parmetros de la distribucin
mu = 170  # Media poblacional
sigma = 8.5  # Desviacin estndar poblacional
n = 25  # Tamao de la muestra
x_bar = 175  # Media muestral a evaluar

# Distribucin muestral de la media
sigma_x_bar = sigma / np.sqrt(n)

# Calcular la probabilidad de que la media muestral sea mayor a 175
z = (x_bar - mu) / sigma_x_bar
probabilidad = 1 - norm.cdf(z)

# Generar datos para la grfica
x = np.linspace(mu - 4 * sigma_x_bar, mu + 4 * sigma_x_bar, 500)
y = norm.pdf(x, loc=mu, scale=sigma_x_bar)

# Crear la grfica
plt.plot(x, y)

# Sombrear el rea para x_bar > 175
x_fill = np.linspace(x_bar, mu + 4 * sigma_x_bar, 500)
y_fill = norm.pdf(x_fill, loc=mu, scale=sigma_x_bar)
plt.fill_between(x_fill, y_fill, color='skyblue', alpha=0.5)

# Agregar lnea vertical en x_bar
plt.axvline(x_bar, color='red', linestyle='--')

# Agregar texto con la probabilidad
plt.text(x_bar + 1, 0.02, f'{probabilidad * 100:.2f}%', fontsize=12, color='red')

# Agregar etiquetas y ttulo
plt.xlabel('Media muestral de las estaturas (cms)')
plt.ylabel('Probabilidad')
plt.title('Probabilidad de que la estatura promedio sea mayor a 175 cms')
plt.grid(True)

# Mostrar la grfica
plt.show()

print(f"La probabilidad de que la estatura promedio de una muestra de 25 estudiantes sea mayor a 175 cms es: {probabilidad * 100:.2f}%")
```
