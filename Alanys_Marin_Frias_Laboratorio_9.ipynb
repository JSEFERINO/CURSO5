```python
# @title **c. Coeficiente de variación - Descriptores numéricos Bivariados 'PROMEDIO vs SEXO'**
print('DEFINITIVA vs SEXO')
df2 = df.groupby('SEXO')['DEFINITIVA'].agg(['mean', 'std', 'var', 'count'])
df2['coef_var(%)'] = round((df2['std'] / df2['mean']) * 100,1)
df2['std_mean'] = round(df2['std'] / np.sqrt(df2['count']),3)
#Applying the calculation to each row and storing as a tuple
df2['IC(μ) - 95%'] = df2.apply(lambda row: (round(row['mean'] - 1.96 * row['std_mean'], 2), round(row['mean'] + 1.96 * row['std_mean'], 2)), axis =1)
df2
```

```python
print('DEFINITIVA vs ESTRATO')
df2 = df.groupby('ESTRATO')['DEFINITIVA'].agg(['mean', 'std', 'var', 'count'])
df2['coef_var(%)'] = round((df2['std'] / df2['mean']) * 100,1)
df2['std_mean'] = round(df2['std'] / np.sqrt(df2['count']),3)
#Applying the calculation to each row and storing as a tuple
df2['IC(μ) - 95%'] = df2.apply(lambda row: (round(row['mean'] - 0.975 * row['std_mean'], 2), round(row['mean'] + 0.975 * row['std_mean'], 2)), axis =1)
df2
```

```python
# prompt: Grafico de medias para los intervalos de confianza

import matplotlib.pyplot as plt

# Assuming df2 is the DataFrame from the previous code block with the calculated confidence intervals
# Example usage with 'DEFINITIVA vs SEXO'
df2 = df.groupby('SEXO')['DEFINITIVA'].agg(['mean', 'std', 'var', 'count'])
df2['coef_var(%)'] = round((df2['std'] / df2['mean']) * 100, 1)
df2['std_mean'] = round(df2['std'] / np.sqrt(df2['count']), 3)
df2['IC(μ) - 95%'] = df2.apply(lambda row: (round(row['mean'] - 1.96 * row['std_mean'], 2), round(row['mean'] + 1.96 * row['std_mean'], 2)), axis=1)

# Extract data for plotting
sexos = df2.index
means = df2['mean']
lower_bounds = [x[0] for x in df2['IC(μ) - 95%']]
upper_bounds = [x[1] for x in df2['IC(μ) - 95%']]

# Create the plot
plt.figure(figsize=(8, 6))
plt.errorbar(sexos, means, yerr=[means - lower_bounds, upper_bounds - means], fmt='o', capsize=5)
plt.xlabel("Sexo")
plt.ylabel("Media de la Definitiva")
plt.title("Medias para los IC DEFINITIVA vs SEXO - Alanys Marin Frias")
plt.grid(True)
plt.show()


# Example usage with 'DEFINITIVA vs ESTRATO'
df2 = df.groupby('ESTRATO')['DEFINITIVA'].agg(['mean', 'std', 'var', 'count'])
df2['coef_var(%)'] = round((df2['std'] / df2['mean']) * 100, 1)
df2['std_mean'] = round(df2['std'] / np.sqrt(df2['count']), 3)
df2['IC(μ) - 95%'] = df2.apply(lambda row: (round(row['mean'] - 0.975 * row['std_mean'], 2), round(row['mean'] + 0.975 * row['std_mean'], 2)), axis=1)

# Extract data for plotting
estratos = df2.index
means = df2['mean']
lower_bounds = [x[0] for x in df2['IC(μ) - 95%']]
upper_bounds = [x[1] for x in df2['IC(μ) - 95%']]

# Create the plot
plt.figure(figsize=(10, 6))
plt.errorbar(estratos, means, yerr=[means - lower_bounds, upper_bounds - means], fmt='o', capsize=5)
plt.xlabel("Estrato")
plt.ylabel("Media de la Definitiva")
plt.title("Medias para los IC DEFINITIVA vs ESTRATO - Alanys Marin Frias")
plt.grid(True)
plt.show()
```

```python
print('DEFINITIVA vs PROGRAMA')
df2 = df.groupby('PROGRAMA')['DEFINITIVA'].agg(['mean', 'std', 'var', 'count'])
df2['coef_var(%)'] = round((df2['std'] / df2['mean']) * 100,1)
df2['std_mean'] = round(df2['std'] / np.sqrt(df2['count']),3)
df2['critical_t'] = df2.apply(lambda row: stats.t.ppf(0.975, df=row['count'] - 1), axis=1)
#Applying the calculation to each row and storing as a tuple
df2['Z - IC(μ) - 95%'] = df2.apply(lambda row: (round(row['mean'] - 1.96 * row['std_mean'], 2), round(row['mean'] + 1.96 * row['std_mean'], 2)), axis =1)
df2['t - IC(μ) - 95%'] = df2.apply(lambda row: (round(row['mean'] - row['critical_t'] * row['std_mean'], 2), round(row['mean'] + row['critical_t'] * row['std_mean'], 2)), axis =1)
df2
```

```python
# prompt: Grafico de medias para los intervalos de confianza con distribución Z y t

# Assuming df2 is the DataFrame from the previous code block with the calculated confidence intervals
# Example usage with 'DEFINITIVA vs PROGRAMA'

# ... (previous code remains the same)


# Extract data for plotting (Z-distribution)
programas = df2.index
means_z = df2['mean']
lower_bounds_z = [x[0] for x in df2['Z - IC(μ) - 95%']]
upper_bounds_z = [x[1] for x in df2['Z - IC(μ) - 95%']]

# Create the plot (Z-distribution)
plt.figure(figsize=(12, 6))
plt.errorbar(programas, means_z, yerr=[means_z - lower_bounds_z, upper_bounds_z - means_z], fmt='o', capsize=5, label='Z-distribution')

# Extract data for plotting (t-distribution)
means_t = df2['mean']
lower_bounds_t = [x[0] for x in df2['t - IC(μ) - 95%']]
upper_bounds_t = [x[1] for x in df2['t - IC(μ) - 95%']]

# Plot t-distribution on the same graph
plt.errorbar(programas, means_t, yerr=[means_t - lower_bounds_t, upper_bounds_t - means_t], fmt='x', capsize=5, label='t-distribution')


plt.xlabel("Programa")
plt.ylabel("Media de la Definitiva")
plt.title("Medias para los IC DEFINITIVA vs PROGRAMA (Z y t) - Alanys Marin Frias")
plt.xticks(rotation=90) # Rotates the x-axis labels for better readability
plt.grid(True)
plt.legend()
plt.tight_layout() # Adjusts the plot layout to prevent labels from overlapping
plt.show()
```
