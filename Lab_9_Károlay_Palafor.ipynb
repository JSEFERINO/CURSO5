```python
print('Definitiva VS Sexo')
df2= df.groupby('SEXO')['DEFINITIVA'].agg(['mean', 'std', 'var', 'count'])
df2 ['coef_var(%)'] = round ((df2['std'] / df2['mean']) *100,1)
df2 ['std_mean'] = round (df2['std'] / np.sqrt(df2['count']),3)
#Applying tbe calculation to each row and storing as a tuple
df2 ['IC(μ)- 95%'] = df2. apply(lambda row: (round(row['mean'], 2), round(row['mean'] + 1.96 * row ['std_mean'],2)), axis=1)
df2
```

```python
print('Definitiva VS Estrato')
df2= df.groupby('ESTRATO')['DEFINITIVA'].agg(['mean', 'std', 'var', 'count'])
df2 ['coef_var(%)'] = round ((df2['std'] / df2['mean']) *100,1)
df2 ['std_mean'] = round (df2['std'] / np.sqrt(df2['count']),3)
#Applying tbe calculation to each row and storing as a tuple
df2 ['IC(μ)- 95%'] = df2.apply(lambda row: (round(row['mean'], 2), round(row['mean'] - 1.96 * row['std_mean'],2), round(row['mean'] + 1.96 * row['std_mean'],2)), axis=1)
degrees_freedom = df2['count'] - 1
alpha = 0.05  # For a 95% confidence interval
t_values = []
for i in range(len(degrees_freedom)):
  t_value = stats.t.ppf(1 - alpha / 2, degrees_freedom.iloc[i])
  t_values.append(t_value)
df2['t-value'] = t_values
df2
```

```python
import matplotlib.pyplot as plt
estratos = df2.index
means = df2['mean']
lower_bounds = [interval[1] for interval in df2['IC(μ)- 95%']]
upper_bounds = [interval[2] for interval in df2['IC(μ)- 95%']]
plt.figure(figsize=(10, 6))
plt.errorbar(estratos, means, yerr=[means - lower_bounds, upper_bounds - means],
             fmt='o', capsize=5, color='green', label="Intervalo de confianza")
plt.xlabel("Estrato")
plt.ylabel("Media de la Definitiva")
plt.title("Intervalo de Confianza de la Definitiva por Estrato-Károlay Palafor")
plt.grid(True)
plt.fill_between(estratos, lower_bounds, upper_bounds, color='pink', alpha=0.3, label="IC 95%")
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```
