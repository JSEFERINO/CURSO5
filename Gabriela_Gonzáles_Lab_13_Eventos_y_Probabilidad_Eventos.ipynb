#  **17. Eventos condicionales y probabilidad Condicional**


## 🎯 **Datos del problema**
### **Una Urna contiene las cantidades de bolas que aparecen a continuación y se extraen tres bolas sin reemplazo**
* Bolas **rojas (R)** = 8
* Bolas **blancas (B)** = 3
* Bolas **azules (A)** = 9
* Total ( N = 20 )
* Se extraen ( n = 3 ) bolas **sin reemplazo**

---

## 🧠 **Eventos a calcular**

a) Las tres sean rojas.
b) Las tres sean blancas.
c) Dos rojas y una blanca.
d) Al menos una sea blanca.
e) Una de cada color.
f) Salgan en el orden rojo, blanco, azul.

---

## 💻 Código en Python

A continuación, todo el código que puedes ejecutar en Jupyter, Colab o cualquier entorno Python.

```python
import math
from scipy.stats import hypergeom

# Datos
N = 20   # total de bolas
R = 8    # rojas
B = 3    # blancas
A = 9    # azules
n = 3    # bolas extraídas

# a) Tres bolas rojas
p_a = hypergeom.pmf(3, N, R, n)

# b) Tres bolas blancas
p_b = hypergeom.pmf(3, N, B, n)

# c) Dos rojas y una blanca
# Para dos rojas: combinaciones de 2 rojas y 1 blanca
p_c = (math.comb(R, 2) * math.comb(B, 1)) / math.comb(N, n)

# d) Al menos una blanca = 1 - P(ninguna blanca)
p_d = 1 - (math.comb(N - B, n) / math.comb(N, n))

# e) Una de cada color
p_e = (math.comb(R, 1) * math.comb(B, 1) * math.comb(A, 1)) / math.comb(N, n)

# f) Orden específico: rojo, blanco, azul (sin reemplazo, ordenado)
# Como el orden importa, se multiplican las probabilidades condicionales
p_f = (R/N) * (B/(N-1)) * (A/(N-2))

# Mostrar resultados
print(f"a) P(3 rojas) = {p_a:.5f}")
print(f"b) P(3 blancas) = {p_b:.5f}")
print(f"c) P(2 rojas y 1 blanca) = {p_c:.5f}")
print(f"d) P(al menos 1 blanca) = {p_d:.5f}")
print(f"e) P(una de cada color) = {p_e:.5f}")
print(f"f) P(roja, blanca, azul en orden) = {p_f:.5f}")
```

## 🌈 Interpretación

* La **mayor probabilidad** ocurre al sacar **una de cada color**, que tiene sentido ya que hay muchas bolas azules y rojas.
* Es **muy poco probable** sacar tres blancas, pues solo hay 3 en total.
* “Al menos una blanca” tiene una probabilidad del 37.7%, lo cual muestra que, aunque pocas, las bolas blancas tienen una presencia moderada.


````python
import math
from scipy.stats import hypergeom

# Datos
N = 20   # total de bolas
R = 8    # rojas
B = 3    # blancas
A = 9    # azules
n = 3    # bolas extraídas

# a) Tres bolas rojas
p_a = hypergeom.pmf(3, N, R, n)

# b) Tres bolas blancas
p_b = hypergeom.pmf(3, N, B, n)

# c) Dos rojas y una blanca
# Para dos rojas: combinaciones de 2 rojas y 1 blanca
p_c = (math.comb(R, 2) * math.comb(B, 1)) / math.comb(N, n)

# d) Al menos una blanca = 1 - P(ninguna blanca)
p_d = 1 - (math.comb(N - B, n) / math.comb(N, n))

# e) Una de cada color
p_e = (math.comb(R, 1) * math.comb(B, 1) * math.comb(A, 1)) / math.comb(N, n)

# f) Orden específico: rojo, blanco, azul (sin reemplazo, ordenado)
# Como el orden importa, se multiplican las probabilidades condicionales
p_f = (R/N) * (B/(N-1)) * (A/(N-2))

# Mostrar resultados
print(f"a) P(3 rojas) = {p_a:.5f}")
print(f"b) P(3 blancas) = {p_b:.5f}")
print(f"c) P(2 rojas y 1 blanca) = {p_c:.5f}")
print(f"d) P(al menos 1 blanca) = {p_d:.5f}")
print(f"e) P(una de cada color) = {p_e:.5f}")
print(f"f) P(roja, blanca, azul en orden) = {p_f:.5f}")
````


# **18. Regla de la Probabilidad Total y Teorema de Bayes**

# **A. Regla de la Probabilidad Total**

### Se considera un experimento que se realiza en dos etapas, en la primera se supone que los posibles sucesos, $A_1, A_2,\ldots,A_n$, constituyen un sistema completo, de tal forma que son conocidas las *probabilidades a priori*, $\mathbb{P}(A_1), \mathbb{P}(A_2),\ldots,\mathbb{P}(A_n)$.

### Si se conocen las probabilidades condicionadas $\mathbb{P}(B\mid A_i)$ para un cierto suceso $B$ y cada $A_i$ se verifica que:    $$\mathbb{P}(B)= \sum _{i=1} ^n \mathbb{P}(B\mid A _i)\mathbb{P}(A _i)$$

## **B. Teorema de Bayes**

### Si conocemos que un suceso esperado $B$ ha ocurrido, nos podríamos preguntar por la probabilidad de ocurrencia de aquellos eventos que positivamente hubiesen originado la ocurrencia de $B$. Tal cálculo es posible usando una expresión que se conoce como la **fórmula de Bayes**
+ $$
\mathbb{P}(A_i|B) = \dfrac{\mathbb{P}(A_i) \mathbb{P}(B|A_i) }{\sum_{j=1}^{k} \mathbb{P}(A_j) \mathbb{P}(B|A_j)}
$$

+ El teorema de Bayes resuelve esta cuestión, llevando el cálculo de las probabilidades a un terreno más natural, expresando las  probabilidades a posteriori $\mathbb{P}(A_i|B)$ en función de las  **verosimilitudes** $\mathbb{P}(B|A_i)$.

## **Solución**.
### **Para la resolución del problema se definen los siguientes sucesos o eventos:**

+ $A=$ \{Enjuiciado cometió el crimen\}
+ $I=$ \{Enjuiciado inocente\}
+ $C=$ \{Enjuiciado culpable\}

### **Probabilidades**:
+ **(a). Hay un asesino entre 5 sospechosos, por tanto, la probabilidad de que una persona elegida al azar sea el asesino es:**
$$\mathbb{P}(A)=\dfrac{1}{5}=0.2, \quad \quad \mathbb{P}(\bar{A})=\dfrac{4}{5}=0.8$$

+ **(b). La probabilidad de creer inocente al verdadero asesino**
 $$\mathbb{P}(I|A)=0.05$$.
 + Además, se sabe que
+  $$\mathbb{P}(C|A)=1-P(I|A)=0.95$$

+ **(c). La probabilidad de creer culpable a una persona inocente**
$$\mathbb{P}(C| \bar{A})=0.08$$
 + Tambien $$\mathbb{P}(I| \bar{A})=0.92$$


### **Solución**
+ **En el problema se pide la probabilidad de que una persona asesina haya sido enjuiciada culpable, es decir, la probabilidad de que una persona sea asesina condicionada a que ha sido enjuicida culpable. Por tanto, la probabilidad requerida es $\mathbb{P}(A|C)$, para calcular dicha probabilidad se recurre al *teorema de Bayes***
+ $$\mathbb{P}(A|C)=\dfrac{\mathbb{P}(C|A)\mathbb{P}(A)}{\mathbb{P}(C|A)\mathbb{P}(A)+\mathbb{P}(C|\bar{A})\mathbb{P}(\bar{A})}=\dfrac{0.95\cdot 0.2}{0.95\cdot 0.2 +0.08 \cdot 0.8}
=\dfrac{0.19}{0.254}=0.748$$

# **19. Regla de la Probabilidad Total y Teorema de Bayes**

## **Ejemplo**. **Supongamos que estamos probando una enfermedad y queremos calcular la probabilidad de que una persona tenga la enfermedad dada una prueba positiva. Para ello, necesitamos los siguientes datos:**

+ **La probabilidad de que una persona tenga la enfermedad** $$P(D)$$
+ **La sensibilidad de la prueba (la probabilidad de que la prueba sea positiva dado que la persona tiene la enfermedad)**, $$P(T∣D)$$
+ **La especificidad de la prueba (la probabilidad de que la prueba sea negativa dado que la persona no tiene la enfermedad,** $$P(\bar{T}∣\bar{D})$$.


+ **Podemos utilizar el teorema de Bayes para calcular la probabilidad de que una persona tenga la enfermedad dado un resultado positivo de la prueba**
$$ P(D∣T)$$:

+ $$\mathbb{P}(D|T)=\dfrac{\mathbb{P}(T|D)\mathbb{P}(D)}{\mathbb{P}(T|D)\mathbb{P}(D)+\mathbb{P}(T|\bar{D})\mathbb{P}(\bar{D})}=\dfrac{0.95\cdot 0.2}{0.95\cdot 0.2 +0.08 \cdot 0.8}
=\dfrac{0.19}{0.254}=0.748$$



```python
def bayes_theorem(prior_prob, sensitivity, specificity):
    # Probabilidad de obtener un resultado positivo de la prueba
    prob_positive_test = (sensitivity * prior_prob) + ((1 - specificity) * (1 - prior_prob))

    # Probabilidad de tener la enfermedad dado un resultado positivo de la prueba
    posterior_prob = (sensitivity * prior_prob) / prob_positive_test

    return posterior_prob

# Datos
prior_prob_disease = 0.20  # Probabilidad de que una persona tenga la enfermedad
sensitivity_test = 0.95     # Sensibilidad de la prueba
specificity_test = 0.92    # EsApecificidad de la prueba

# Calculando la probabilidad posterior usando el teorema de Bayes
posterior_probability = bayes_theorem(prior_prob_disease, sensitivity_test, specificity_test)

print("La probabilidad de que una persona tenga la enfermedad dado un resultado positivo de la prueba es:", posterior_probability)


````python
def bayes_theorem(prior_prob, sensitivity, specificity):
    # Probabilidad de obtener un resultado positivo de la prueba
    prob_positive_test = (sensitivity * prior_prob) + ((1 - specificity) * (1 - prior_prob))

    # Probabilidad de tener la enfermedad dado un resultado positivo de la prueba
    posterior_prob = (sensitivity * prior_prob) / prob_positive_test

    return posterior_prob

# Datos
prior_prob_disease = 0.20  # Probabilidad de que una persona tenga la enfermedad
sensitivity_test = 0.95     # Sensibilidad de la prueba
specificity_test = 0.92    # EsApecificidad de la prueba

# Calculando la probabilidad posterior usando el teorema de Bayes
posterior_probability = bayes_theorem(prior_prob_disease, sensitivity_test, specificity_test)

print("La probabilidad de que una persona tenga la enfermedad dado un resultado positivo de la prueba es:", posterior_probability)
````

El video es un laboratorio de estadística que usa un programa de computadora (Python en Google Colab) para demostrar cómo funciona el azar. Se enfoca en una idea clave llamada la Ley de los Grandes Números. En palabras simples, esta ley nos dice que, aunque no podemos saber qué pasará en un solo intento (como un solo lanzamiento de un dado), sí podemos predecir lo que pasará si repetimos ese intento muchísimas veces.

La Ley de los Grandes Números (LGN)

El experimento principal que se demuestra es el de lanzar un dado:

1. Pocos Lanzamientos: Si lanzamos un dado solo 200 veces, los resultados (cuántas veces sale el 1, el 2, el 3, etc.) son muy irregulares y diferentes entre sí.

2. Muchos Lanzamientos: Pero, si lanzamos el dado 200,000 veces, el número de veces que sale cada cara (1, 2, 3, 4, 5 y 6) se vuelve casi idéntico. Como hay 6 caras, la probabilidad "correcta" es que cada una salga 1 de cada 6 veces.

La Conclusión de la LGN: La estadística demuestra que cuantas más veces repitas un experimento, el resultado real (la frecuencia observada) se acercará más a la probabilidad teórica o "verdadera" que esperas. Lo mismo se demuestra al lanzar una moneda, donde la cantidad de caras y sellos se acerca al 50% al repetir miles de veces el lanzamiento.
El laboratorio no solo prueba que el azar sigue reglas, sino que también muestra cómo la estadística se usa para problemas más complicados:

• Urnas y Colores: Se simula sacar bolas de una urna con colores. Al hacerlo muchas veces, las bolas que se sacan de cada color coinciden con el porcentaje que deberían tener (por ejemplo, si el 50% son azules, casi el 50% de las extracciones serán azules).

• Cálculos Condicionales: Al final, el video usa el programa para resolver rápidamente problemas de Probabilidad Condicional y el Teorema de Bayes. Esto se usa en la vida real para cosas como calcular la probabilidad de tener una enfermedad si tu prueba da positivo, usando toda la información disponible para hacer una predicción más precisa.

En resumen, el video nos enseña que el azar no es totalmente caótico. La Ley de los Grandes Números es la regla que le da orden. Gracias a esta ley, podemos usar la estadística para predecir el futuro con bastante certeza, siempre y cuando tengamos muchos datos y repitamos los experimentos suficientes veces.
