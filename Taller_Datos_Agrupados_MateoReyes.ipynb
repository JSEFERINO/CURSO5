#Punto 8: Medidas de tendencia central agrupadas: Media, Mediana y Moda

```python
import numpy as np
import pandas as pd

edades = [
  19,18,18,21,18,18,18,17,18,22,18,18,18,18,18,22,21,19,18,19,18,19,18,17,
  19,20,19,18,17,18,20,19,20,22,18,19,19,18,19,20,19,18,17,19,19,19,19,18,
  17,18,19,20,23,18,18,20,18,18,18,18,18,19,17,17,20,19,18,19,19,21,19,21,
  17,19,18,18,17,17,18,18,17,18,19,17,18,18,18,18,20,18,18,18,17,19,18,17,
  18,18,18,18,18,18,18,18,19,18,19,18,18,17,18,18,18,18,18,18,18,18,24,18,
  19,19,18,17,17,17,20,17,18,17,18,18,19,18,18,18
]

num_clases = int(np.ceil(1 + 3.322 * np.log10(len(edades))))
max_val = max(edades)
min_val = min(edades)
amplitud = np.ceil((max_val - min_val) / num_clases)
intervalos = np.arange(min_val, max_val + amplitud, amplitud)
frecuencias, _ = np.histogram(edades, bins=intervalos)

marcas_clase = [(intervalos[i] + intervalos[i+1]) / 2 for i in range(len(intervalos)-1)]
frecuencia_acumulada = np.cumsum(frecuencias)

media = sum(marcas_clase[i] * frecuencias[i] for i in range(len(marcas_clase))) / sum(frecuencias)

n = sum(frecuencias)
pos_mediana = n / 2

for i, fa in enumerate(frecuencia_acumulada):
    if fa >= pos_mediana:
        L = intervalos[i]
        F = frecuencia_acumulada[i-1] if i > 0 else 0
        f = frecuencias[i]
        mediana = L + ((pos_mediana - F) / f) * amplitud
        break

moda_index = np.argmax(frecuencias)
L_m = intervalos[moda_index]
f_m = frecuencias[moda_index]
f_m1 = frecuencias[moda_index - 1] if moda_index > 0 else 0
f_m2 = frecuencias[moda_index + 1] if moda_index < len(frecuencias) - 1 else 0
moda = L_m + ((f_m - f_m1) / ((f_m - f_m1) + (f_m - f_m2))) * amplitud

print(f"Media: {media}")
print(f"Mediana: {mediana}")
print(f"Moda: {moda}")
```

#Punto 9: Halle medidas de posición para datos agrupados: Q1, Q3, D3, D8 e interprete los valores encontrados

```python
import numpy as np
import pandas as pd

edades = [
  19,18,18,21,18,18,18,17,18,22,18,18,18,18,18,22,21,19,18,19,18,19,18,17,
  19,20,19,18,17,18,20,19,20,22,18,19,19,18,19,20,19,18,17,19,19,19,19,18,
  17,18,19,20,23,18,18,20,18,18,18,18,18,19,17,17,20,19,18,19,19,21,19,21,
  17,19,18,18,17,17,18,18,17,18,19,17,18,18,18,18,20,18,18,18,17,19,18,17,
  18,18,18,18,18,18,18,18,19,18,19,18,18,17,18,18,18,18,18,18,18,18,24,18,
  19,19,18,17,17,17,20,17,18,17,18,18,19,18,18,18
]

num_clases = int(np.ceil(1 + 3.322 * np.log10(len(edades))))
max_val = max(edades)
min_val = min(edades)
amplitud = np.ceil((max_val - min_val) / num_clases)
intervalos = np.arange(min_val, max_val + amplitud, amplitud)
frecuencias, _ = np.histogram(edades, bins=intervalos)

frecuencia_acumulada = np.cumsum(frecuencias)
n = sum(frecuencias)

percentiles = {"Q1": 0.25, "Q3": 0.75, "D3": 0.3, "D8": 0.8}
resultados = {}

for nombre, p in percentiles.items():
    pos = p * n
    for i, fa in enumerate(frecuencia_acumulada):
        if fa >= pos:
            L = intervalos[i]
            F = frecuencia_acumulada[i - 1] if i > 0 else 0
            f = frecuencias[i]
            resultados[nombre] = L + ((pos - F) / f) * amplitud
            break

for nombre, valor in resultados.items():
    print(f"{nombre}: {valor}")
```

#Punto 10: Construya un intervalo de Confianza de 95% para la media real de las edades e interprete el resultado.

```python
import numpy as np
import scipy.stats as stats

edades = [
  19,18,18,21,18,18,18,17,18,22,18,18,18,18,18,22,21,19,18,19,18,19,18,17,
  19,20,19,18,17,18,20,19,20,22,18,19,19,18,19,20,19,18,17,19,19,19,19,18,
  17,18,19,20,23,18,18,20,18,18,18,18,18,19,17,17,20,19,18,19,19,21,19,21,
  17,19,18,18,17,17,18,18,17,18,19,17,18,18,18,18,20,18,18,18,17,19,18,17,
  18,18,18,18,18,18,18,18,19,18,19,18,18,17,18,18,18,18,18,18,18,18,24,18,
  19,19,18,17,17,17,20,17,18,17,18,18,19,18,18,18
]

confianza = 0.95
media = np.mean(edades)
desv_est = np.std(edades, ddof=1)
n = len(edades)

error_estandar = desv_est / np.sqrt(n)
intervalo = stats.t.interval(confianza, df=n-1, loc=media, scale=error_estandar)

print(f"Intervalo de confianza del 95%: ({intervalo[0]:.5f}, {intervalo[1]:.5f})")
```
