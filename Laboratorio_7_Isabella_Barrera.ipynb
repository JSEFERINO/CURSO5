```python
# LABORATORIO 7 - MUESTREO DE VARIANZAS
# Profesor Julio Hurtado
# Estudiante: Isabella Barrera Babilonia

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# PARTE 1: VARIANZAS DE EDADES DE ESTUDIANTES

np.random.seed(42)
poblacion_edades = np.random.randint(18, 24, size=1000)
tamano_muestra = 5
repeticiones = 100

varianzas_edades = []
for _ in range(repeticiones):
    muestra = np.random.choice(poblacion_edades, tamano_muestra, replace=True)
    varianzas_edades.append(np.var(muestra, ddof=1))

df_edades = pd.DataFrame({'Varianzas': varianzas_edades})
display(df_edades.head())

plt.figure(figsize=(8, 6))
plt.hist(df_edades['Varianzas'], bins=10, edgecolor='black', color='lightblue')
plt.title('Histograma del Muestreo de Varianzas - Edades', fontsize=14)
plt.xlabel('Varianza')
plt.ylabel('Frecuencia')
plt.grid(True, alpha=0.3)
plt.show()

# PARTE 2: VARIANZAS DE LANZAMIENTOS DE DADOS

poblacion_dados = np.arange(1, 7)
tamano_muestra_dados = 5
repeticiones_dados = 200

varianzas_dados = []
for _ in range(repeticiones_dados):
    muestra_dados = np.random.choice(poblacion_dados, tamano_muestra_dados, replace=True)
    varianzas_dados.append(np.var(muestra_dados, ddof=1))

df_dados = pd.DataFrame({'Varianzas': varianzas_dados})
display(df_dados.head())

plt.figure(figsize=(8, 6))
plt.hist(df_dados['Varianzas'], bins=10, color='lightblue', edgecolor='black')
plt.title('Histograma del Muestreo de Varianzas - Lanzamientos de Dados', fontsize=14)
plt.xlabel('Varianza')
plt.ylabel('Frecuencia')
plt.grid(True, alpha=0.3)
plt.show()
```
